```python
num = range(0,1000)
for i in num:
    if i > 1:
        for y in range(2,i): 
            if (i % y) == 0: 
                print(i, "is not a prime number") 
                break
            else: 
                print(i, "is a prime number") 
    else: 
        print(i, "is not a prime number") 
```

```python
NO_OF_CHARS = 256
def canFormPalindrome(string): 
    count = [0 for i in range(NO_OF_CHARS)] 
    for i in string: 
        count[ord(i)] += 1
    odd = 0
    for i in range(NO_OF_CHARS): 
        if (count[i] & 1): 
            odd += 1   
        if (odd > 1): 
            return False  
    return True
if(canFormPalindrome("tttttttttttt")): 
    print("Yes") 
else: 
    print("No")
if(canFormPalindrome("tototpyo")): 
    print("Yes") 
else: 
    print("NO") 
```
